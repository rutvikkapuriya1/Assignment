```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel(r'C:\Users\rutvi\Downloads\rawdata.xlsx')
df
```

```python
df['position'] = df['position'].str.lower()
```

```python
df.dtypes
```

```python
# Convert 'time' column to datetime format
df['time'] = pd.to_datetime(df['time'], format='%H:%M:%S')
df
```

```python
df = df.sort_values(by='time')
df
```

```python
df['duration'] = df.groupby(['date'])['time'].diff().fillna(pd.Timedelta(seconds=0))
df.head(5)
```

```python
activities_count = df.groupby(['date', 'activity']).size().unstack(fill_value=0).reset_index()
activities_count
```

```python
duration_sum = df.groupby(['date', 'position'])['duration'].sum().unstack(fill_value=pd.Timedelta(seconds=0)).reset_index()
duration_sum
```

```python
output = pd.merge(activities_count, duration_sum, on='date', how='outer')
output
```

```python
output.rename(columns={'picked': 'pick_activities', 'placed': 'place_activities', 'inside': 'inside_duration', 'outside': 'outside_duration'}, inplace=True)
output
```
